## (1) Sales by city

```python
# Make DataFrame Pipeline Interactive
idf = df.interactive()
```

```python
idf.info()
```

```python
# Define Panel widgets
values = (dt.datetime(2021,1, 1, 00, 00), dt.datetime(2021, 12, 31, 00, 00))

date_range = pn.widgets.DateRangeSlider(
                name='date_range', 
                start=dt.datetime(2021, 1, 1, 00, 00),end=dt.datetime(2021, 12, 31, 00, 00),
                value=values)
```

```python
date_range
```

```python
#Ratio button for sales measures
#yaxis_sales = pn.widgets.RadioButtonGroup(
    #name='Y axis', 
    #options=['total', 'total_per_city',],
    #button_type='success'
#)
```

```python
#yaxis_sales
```

```python
citys = ['Yangon','Naypyitaw','Mandalay'] 
sales_pipeline = (
    idf[
        (idf.created_time >= date_range.start) & (idf.created_time <= date_range.end) &
        (idf.city.isin(citys))
    ]
    .groupby(['city', 'created_time'])['total'].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='created_time')  
    .reset_index(drop=True)
)
```

```python
sales_pipeline
```

```python

sales_plot = sales_pipeline.hvplot(x = 'created_time', by = 'city',y='total',kind = 'bar',stacked = True, title="Sales by city")
```

```python
sales_plot
```

## (2) chart 2: sales by hours

```python
sales_by_hour_pipeline = (
    idf[
        (idf.created_time >= date_range.start) & (idf.created_time <= date_range.end) 
    ]
    .groupby(['hours'])['total'].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='hours')  
    .reset_index(drop=True)
)
```

```python
sales_by_hour_pipeline
```

## (3) Chart 3: unit price  and gross income - scatter plot

```python
total_and_income_pipeline = (
    idf[
        (idf.created_time >= date_range.start) & (idf.created_time <= date_range.end) 
    ]
    .groupby(['invoice_id','gross income','city','unit price'])['total'].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='invoice_id')  
    .reset_index(drop=True)
)
```

```python
total_and_income_pipeline
```

```python
total_and_income_pipeline_chart = total_and_income_pipeline.hvplot(x ='gross income',y='unit price',kind = 'scatter',title="unit price and gross income scatterplot")
```

```python
total_and_income_pipeline_chart 
```

## (4) Chart 4: Sales table by city

```python
sales_table_pipeline = (
    idf[
        (idf.created_time >= date_range.start) & (idf.created_time <= date_range.end) 
    ]
    .groupby(['city','customer_type','gender'])['total'].sum()
    .to_frame()
    .reset_index()
    .sort_values(by='city')  
    .reset_index(drop=True)
)
```

```python
sales_table_pipeline
```

```python
sales_table = sales_table_pipeline.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 12, sizing_mode='stretch_width')
```

```python
sales_table 
```
